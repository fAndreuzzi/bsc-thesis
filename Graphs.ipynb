```python
import networkx as nx
import networkx.drawing.nx_pylab as nx_draw
from networkx.drawing.nx_agraph import to_agraph

from IPython.display import Image
import os
```

## Tree graph

```python
graph = nx.balanced_tree(2, 3, create_using=nx.DiGraph)

for node in graph.nodes:
    graph.nodes[node]['shape'] = 'circle'
    graph.nodes[node]['fixedsize'] = True
    graph.nodes[node]['width'] = 0.3

for edge in graph.edges:
    graph.edges[edge]['arrowsize'] = 0.5

agraph = to_agraph(graph)
agraph.layout('dot')
agraph.graph_attr.update(dpi=300)
agraph.draw('test.png')

img = Image(filename='test.png', width = 800, height = 400)
os.remove('test.png')

img
```

## Hopcroft

```python
n = 3

graph = nx.DiGraph()
graph.add_nodes_from(range(1,n+1))
graph.add_edges_from([(i, n // 2 + 2 * i - 1) for i in range(1, n+1)])
graph.add_edges_from([(n//2+i, 2*i-1) for i in range(n//2+1, n+1)])
graph.add_edges_from([(n//4+1, 2*i-1) for i in range(1,n//4+1)])

for node in graph.nodes:
    graph.nodes[node]['shape'] = 'circle'
    graph.nodes[node]['fixedsize'] = True
    graph.nodes[node]['width'] = 0.3

for edge in graph.edges:
    graph.edges[edge]['arrowsize'] = 0.5

agraph = to_agraph(graph)
agraph.layout('dot')
agraph.graph_attr.update(dpi=300)
agraph.draw('test.png')

img = Image(filename='test.png', width = 100, height = 100)
os.remove('test.png')

img
```
